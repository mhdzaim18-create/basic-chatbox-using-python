```python
def Zaim_chatbot():
    print("--- Zaim Rule-Based Chatbot ---")
    print("Type 'bye' to exit the conversation.")
    
    while True:
        # Get user input and convert to lowercase for easier matching
        user_input = input("You: ").lower().strip()

        # Rule-based logic for responses
        if user_input in ["hello", "hi", "hey"]:
            print("Chatbot: Hi! How can I help you today?")
        
        elif user_input in ["how are you", "how's it going"]:
            print("Chatbot: I'm fine, thanks for asking!")
            
        elif user_input in ["what is your name", "who are you"]:
            print("Chatbot: I am the Zaim Basic Chatbot.")
            
        elif user_input == "bye":
            print("Chatbot: Goodbye! Have a great day!")
            break # Exit the loop
            
        else:
            print("Chatbot: I'm sorry, I don't understand that. Try saying 'hello' or 'how are you'.")

if __name__ == "__main__":
    Zaim_chatbot()
```
